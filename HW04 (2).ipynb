To create a 95% confidence inerval using the standard error of a bootstrapped dataset, we should first calculate the standard deviation of the bootstrapped dataset to give us the standard error. We also calculate the mean of the bootstrapped dataset (i.e. the mean of means). We then multiply the standard error by 1.96 (the Z-Score for a 95% confidence interval) to get the margin of error (MEM), which is how accurate our mean is. Our 95% confidence interval is one MEM from the mean of the bootstrapped dataset (i.e. one MEM to the right and left of the mean).

Q3

To create an exact 95% confidence interval without using the standard deviation of the bootstrapped dataset, we could find the 97.5th and 2.5th percentile of the bootstrapped dataset. The 2.5th percentile would be the lower bound of the confidence interval and the 97.5th percentile would be the upper bound. The 2.5th percentile means that 2.5% of the data lies below it, and the 97.5th percentile means that 2.5% of the data lies above it. This means that 95% of the data lies between the 97.5th percentile and the 2.5th percentile. Using these two percentiles as the lower and upper bounds of our interval will guarantee (with 95% accuracy) that the actual mean lies between that interval. To calculate those percentiles, we can use the numpy.quantile() method imported from the numpy library on the bootstrapped dataset. The function takes the dataset and the percentiles we require as parameters. So in this case, the dataset would be the bootstrapped data set and we would use the function to calculate the 2.5th percentile and the 97.5th percentile to use as the lower and upper bounds of our interval.


````python
#Q4

import numpy as np

#Generate a random dataset by fixing the seed.
np.random.seed(42) 
original_data = np.random.normal(loc=50, scale=10, size=100)  # Sets the data set so it has a mean of 50, standard deviation of 10 and a sample size of 100.


n_bootstrap_samples = 1000 #To generate 1000 bootstrapped samples.
bootstrap_means = [] #Empty list that will eventually contain the means of each bootstrapped sample.

# Generating 1000 bootstrap samples and calculating their mean then storing it in the bootstrap_medians list.
for _ in range(n_bootstrap_samples): #For loop that loops 1000 times.
    bootstrap_sample = np.random.choice(original_data, size=len(original_data), replace=True) #Generates the bootstrap sample.
    bootstrap_means.append(np.mean(bootstrap_sample)) #Calculates the mean of the bootstrap sample and appends it to the bootstrap_means list.

#Calculating the 95% confidence interval using the percentiles we discussed in the previous question.
lower_bound = np.quantile(bootstrap_means, 0.025) #Lower bound of the 95% confidence interval (2.5th percentile)
upper_bound = np.quantile(bootstrap_means, 0.975) #Upper bound of the 95% confidence interval (97.5th percentile)

print(f"Original Data Mean: {np.mean(original_data):.2f}") #Prints the mean of the original dataset
print(f"95% Bootstrapped Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})") #Prints the 95% confidence interval.
````

### Summary of Interaction on Bootstrapping, Standard Deviation, and Confidence Intervals

**1. Standard Deviation and Standard Error of the Mean:**
- **Standard Deviation** measures the variability or spread of a set of data points around the mean.
- **Standard Error of the Mean (SEM)** measures how much the sample mean is expected to vary from the true population mean. It is calculated by dividing the standard deviation by the square root of the sample size.

**2. Creating a 95% Confidence Interval:**
- To create a 95% confidence interval using the SEM:
  1. Calculate the sample mean (\(\bar{x}\)).
  2. Calculate the SEM.
  3. Determine the Z-score for 95% confidence (approximately 1.96).
  4. Calculate the margin of error (ME) as \(1.96 \times \text{SEM}\).
  5. Construct the confidence interval as \((\bar{x} - \text{ME}, \bar{x} + \text{ME})\).

**3. Bootstrapped Confidence Interval:**
- To create a 95% bootstrapped confidence interval using the bootstrapped means:
  1. Generate multiple bootstrap samples from the original dataset.
  2. Calculate the mean for each bootstrap sample.
  3. Use `np.quantile` to find the 2.5th and 97.5th percentiles of the bootstrap means, which gives the lower and upper bounds of the 95% confidence interval.

**4. Python Code Example:**
- We provided a Python code example to generate a sample dataset, create bootstrap samples, and calculate the 95% confidence interval using NumPy:

```python
import numpy as np

# Step 1: Create a sample dataset
np.random.seed(42)
original_data = np.random.normal(loc=50, scale=10, size=100)

# Step 2: Generate bootstrap samples and calculate their means

    bootstrap_sample = np.random.choice(original_data, size=len(original_data), replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))

# Step 3: Calculate the 95% confidence interval using np.quantile

upper_bound = np.quantile(bootstrap_means, 0.975)

```

1) Bootstrapping starts by calculating the average value of a given dataset that we have. Suppose we wanted to measure the average beak length of ducks. Since it is impossible to measure and record the beak length of every single duck in the world, we will take a relatively small sample of ducks and measure their beak lengths. Suppose we choose 10 ducks. After calculating the average beak length of the 10 ducks we had, we then create a new dataset that has the same amount of values as the original dataset (in our case, 10). We randomly choose 10 beak lengths from the original sample and plot them in a new dataset, except we can choose the same duck length more than once. Essentially, we're creating a new sample using the same values we already measured. We then calculate and record the average of the new dataset we "created". We repeat this process several times. We then plot all the average beak lengths we collected on a histogram. A histogram is basically a visualization of all the averages we collected grouped together in "bins". For example, if we had 15 average beak lengths that lie within 40-50cm, we could group all of these averages together in one bin of width of 10 (since our range is 40-50) and height 15 (since we had 15 values between 40 and 50 cm). We call the new histogram of averages the bootstrapped dataset.


2) The main purpose of bootstrapping is to see how accurate the average beak length that we originally calculated from a sample of 10 ducks was to the value of the true average beak length (had we measured the beak length of every single duck in the world). Bootstrapping doesn't tell us exactly what the true average is, but it tells us how close our value was to it.


3) From the bootstrapped dataset, we can construct a 95% confidence interval. A 95% confidence interval is a range of values where, if we were to retake the original samples several times, the true value of the population average would be in the interval we constructed 95% of the time. The confidence interval allows us to conclude with 95% certainty that the true value of the average lies in that interval. So we can verify the hypothetical mean we calculated by seeing if it lies in that interval or not.

Q8

We want to determine whether or not the vaccine produced by AliTech is effective. To do this, we must inspect whether the Null Hypothesis can be rejected or not. The Null Hypothesis states that the population average of the difference between the final health score and intial health score (i.e. the average of the health scores difference if the vaccine were to be tested on every single person with the disease) can be 0. Since testing the vaccine on the entire population would be practically impossible, we can use bootstrapping instead to simulate testing the vaccine on the population using only the original sample we took.




````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("patients.csv")

# Calculate the difference in health scores
df['HealthScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Calculate the mean of the HealthScoreDifference
mean_diff = df['HealthScoreDifference'].mean()

# Calculate the frequency of each bin
counts, bins = np.histogram(df['HealthScoreDifference'], bins=11)

# Plot the bar plot
plt.figure(figsize=(10, 6))
plt.bar(bins[:-1], counts, width=np.diff(bins), edgecolor='k', alpha=0.7, align='edge')

# Add a vertical line at the mean
plt.axvline(mean_diff, color='r', linestyle='dashed', linewidth=1)

# Add text for the mean value
plt.text(mean_diff + 0.1, max(counts) / 2, f'Mean: {mean_diff:.2f}', color='r')

# Add titles and labels
plt.title('Histogram of Health Score Differences')
plt.xlabel('Health Score Difference')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
````

Above is a representation of the health score difference collected from the 9 patients. None of the patients observed no change in their health score. The mean health score difference of this sample is approximately 3.30.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to perform bootstrapping with a fixed seed
def bootstrap(data, n_iterations, seed):
    np.random.seed(seed)
    bootstrapped_means = []
    for _ in range(n_iterations):
        # Resample with replacement
        sample = data.sample(frac=1, replace=True)
        # Calculate the mean of the HealthScoreDifference
        mean_diff = sample['HealthScoreDifference'].mean()
        bootstrapped_means.append(mean_diff)
    return bootstrapped_means

# Load the original dataset
df = pd.read_csv("patients.csv")

# Calculate the difference in health scores
df['HealthScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Perform bootstrapping with 1000 iterations and a fixed seed
n_iterations = 1000
seed = 44
bootstrapped_means = bootstrap(df, n_iterations, seed)

# Convert the bootstrapped means to a DataFrame
bootstrapped_df = pd.DataFrame(bootstrapped_means, columns=['BootstrappedMeanDifference'])

# Save the bootstrapped dataset to a CSV file
bootstrapped_df.to_csv('bootstrapped_means.csv', index=False)

# Calculate the mean of the bootstrapped means
mean_bootstrapped = bootstrapped_df['BootstrappedMeanDifference'].mean()

# Plot the histogram for the bootstrapped means
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(bootstrapped_df['BootstrappedMeanDifference'], bins=30, edgecolor='k', alpha=0.7)

# Add a vertical line at the mean of the bootstrapped means
plt.axvline(mean_bootstrapped, color='r', linestyle='dashed', linewidth=1)

# Add text for the mean value
plt.text(mean_bootstrapped + 0.1, max(counts) / 2, f'Mean: {mean_bootstrapped:.2f}', color='r')

# Add titles and labels
plt.title('Histogram of Bootstrapped Mean Differences')
plt.xlabel('Bootstrapped Mean Difference')
plt.ylabel('Frequency')
plt.grid(True)


# Show the plot
plt.show()
````

The histogram above represents the mean difference in health scores for 1000 bootstrapped simulations (i.e. the original sample was bootstrapped 1000 times and the mean health difference was calculated each time and plotted on the figure above). The mean (of means) is observed to be approximately 3.31. We can see that some of the 1000 samples had a mean of 0. To determine whether we can reject the Null Hypothesis or not, we must construct a confidence interval. 

````python
# Calculate the mean of the bootstrapped means
mean_bootstrapped = bootstrapped_df['BootstrappedMeanDifference'].mean()

# Calculate the 2.5th and 97.5th percentiles
lower_bound = np.percentile(bootstrapped_df['BootstrappedMeanDifference'], 2.5)
upper_bound = np.percentile(bootstrapped_df['BootstrappedMeanDifference'], 97.5)

# Plot the histogram for the bootstrapped means
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(bootstrapped_df['BootstrappedMeanDifference'], bins=30, edgecolor='k', alpha=0.7)

# Add a vertical line at the mean of the bootstrapped means
plt.axvline(mean_bootstrapped, color='r', linestyle='dashed', linewidth=1)
plt.text(mean_bootstrapped + 0.1, max(counts) / 2, f'Mean: {mean_bootstrapped:.2f}', color='r')

# Add vertical lines at the 2.5th and 97.5th percentiles
plt.axvline(lower_bound, color='b', linestyle='dashed', linewidth=1)
plt.axvline(upper_bound, color='b', linestyle='dashed', linewidth=1)
plt.text(lower_bound - 0.8, max(counts) / 2, f'2.5th: {lower_bound:.2f}', color='b')
plt.text(upper_bound + 0.1, max(counts) / 2, f'97.5th: {upper_bound:.2f}', color='b')


# Add titles and labels
plt.title('Histogram of Bootstrapped Mean Differences with 95% Confidence Interval')
plt.xlabel('Bootstrapped Mean Difference')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
````

The histogram has been modified to add blue vertical lines at the 2.5th percentile and the 97.5th percentile. Between these two blue lines, 95% of the means lie. Within this interval, we can conclude with 95% certainty that the the true population mean lies within this interval. We can also conclue that 0 is not in this interval, which means, according to the 95% confidence interval we created, we can reject the Null Hypothesis with 95% confidence.

Conclusion:

We can conclude with 95% confidence that the vaccine does indeed work since we concluded (with 95% confidence) that we can reject the Null Hypothesis, meaning we reject the case of the population average of the difference of health scores being 0.
